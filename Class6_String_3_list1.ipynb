```python
# isupper() - it returns True if the char in string are in upper case else will return False
s1 = "abc" # False
s2 = "ABC" # True
s3 = "*" # False
s4 = "2004"# False
s5 = "ABCabc12" # False
print(s1.isupper())
print(s2.isupper())
print(s3.isupper())
print(s4.isupper())
print(s5.isupper())
```

```python
# islower() - it returns True if the char in string are in lower case else will return False
s1 = "abc" # True
s2 = "ABC" # False
s3 = "*" # False
s4 = "2004"# False
s5 = "ABCabc12" # False
print(s1.islower())
print(s2.islower())
print(s3.islower())
print(s4.islower())
print(s5.islower())
```

```python
# istitle() - it returns True if the char in string are in lower case else will return False
s1 = "abc" # False
s2 = "ABC" # False
s3 = "*" # False
s4 = "2004"# False
s5 = "ABCabc12" # False
s6 = "Hello World" # True
print(s1.istitle())
print(s2.istitle())
print(s3.istitle())
print(s4.istitle())
print(s5.istitle())
print(s6.istitle())
```
