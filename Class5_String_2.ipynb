```python
Syntax
            var[start:stop:step]
```

```python
if the step is not mentioned the default 
step size ----> 1 ( L to R)
```

```python
# lion
a[0:4:1]
```

```python
# ion
a[-11:-8:1]
# -11 -8 ----> L to R
# 1 ---------> L to R
```

```python
# ion
a[-11:-8:-1]
# -11 -8 ----> L to R
# -1 ---------> R to L
```

```python
# messi
a[7:12:-1]
```

```python
# issem
a[11:6:]
# 11 6 ----> R L
# 1 ----> L R
```

```python
# issem
a[11:6:-1]
# 11 6 ----> R L
# -1 ----> R L
```

```python
a[::]
# 0
# end of the str
# 1
```

```python
a[0:12:1]
```

```python
a[0:12:2]
```

```python
a[0:12:3]
```

```python
# reverse a string
a[::1]
# 0
# end 
# 1
```

```python
a[::-1]
# -1
# end 
# -1
```

```python
a[-1:-13:-2]
# -1
# end 
# -1
```

```python
a[-1::-2]
# -1
# end 
# -1
```

```python
# isalnum - it returns True if the string is either an alphabet or a number else will return False
s1 = "abc" # True
s2 = "123abc" # True
s3 = "*" # False
s4 = "2004"# True
print(s1.isalnum())
print(s2.isalnum())
print(s3.isalnum())
print(s4.isalnum())
```

```python
# isalpha - it returns True if the string is  an alphabet else will return False
s1 = "abc" # True
s2 = "123abc" # False
s3 = "*" # False
s4 = "2004"# False
print(s1.isalpha())
print(s2.isalpha())
print(s3.isalpha())
print(s4.isalpha())
```

```python
# isnumeric - it returns True if the string is  a number else will return False
s1 = "abc" # False
s2 = "123abc" # False
s3 = "*" # False
s4 = "20040090"# True
print(s1.isnumeric())
print(s2.isnumeric())
print(s3.isnumeric())
print(s4.isnumeric())
```

```python
# isascii- it returns True is the string is having an ASCII value
s1 = "abc" # True
s2 = "123abc" # True
s3 = "*" # True
s4 = "20040090"# True
s5 = "" # False
print(s1.isascii())
print(s2.isascii())
print(s3.isascii())
print(s4.isascii())
print(s5.isascii())
```
